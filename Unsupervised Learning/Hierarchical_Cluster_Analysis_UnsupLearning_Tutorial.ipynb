```python
from scipy.cluster.hierarchy import linkage

import pandas as pd
import numpy as np

df = pd.read_csv("/content/USArrests.csv",index_col=0)

hc_complete = linkage(df,"complete")  # we'll use and compare these 2 dendrogram creation methods
hc_average = linkage(df,"average")
```

```python
from scipy.cluster.hierarchy import dendrogram
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.title("Hierarchical Clustering Dendrogram")
plt.ylabel("Distances")
plt.xlabel("Samples (States)")
dendrogram(hc_complete,leaf_font_size=10);
```

```python
plt.figure(figsize=(10,5))
plt.title("Hierarchical Clustering Dendrogram")
plt.ylabel("Distances")
plt.xlabel("Samples (States)")
dendrogram(hc_complete,
           truncate_mode="lastp",
           p=4,                              # We can specify how many cluster we want to divide our data into bu using "truncate_mode" and "p" argument
           leaf_font_size=10);
```

```python
#Creating Dendrogram by "complete" method

plt.figure(figsize=(10,5))
plt.title("Hierarchical Clustering Dendrogram")
plt.ylabel("Distances")
plt.xlabel("Samples (States)")
dendrogram(hc_complete,
           truncate_mode="lastp",
           p=8,                              # We can specify how many cluster we want to divide our data into bu using "truncate_mode" and "p" argument
           leaf_font_size=10);
```

**COMPARING "COMPLETE" & "AVERAGE" LINKAGE METHODS

```python
#Creating Dendrogram by "complete" method

plt.figure(figsize=(10,5))
plt.title("Hierarchical Clustering Dendrogram-Complete Method")
plt.ylabel("Distances")
plt.xlabel("Samples (States)")
dendrogram(hc_complete,leaf_font_size=10);
```

```python
#Creating Dendrogram by "average" method

plt.figure(figsize=(10,5))
plt.title("Hierarchical Clustering Dendrogram-Average Method")
plt.ylabel("Distances")
plt.xlabel("Samples (States)")
dendrogram(hc_average,leaf_font_size=10);
```
