```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale,StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.metrics import confusion_matrix,accuracy_score,mean_squared_error,r2_score,roc_auc_score,roc_curve,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

import warnings
warnings.filterwarnings('ignore',category=DeprecationWarning)
warnings.filterwarnings('ignore',category=FutureWarning)
data = pd.read_csv("diabetes.csv")
df = pd.DataFrame(data)

y = df['Outcome']
X = df.drop(columns = 'Outcome')
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30,random_state = 42)
```

```python
mlp_classifier = MLPClassifier().fit(X_train,y_train)
```

```python
mlp_classifier.coefs_
```

```python
?mlp_classifier
```

```python
y_pred = mlp_classifier.predict(X_test)
```

```python
accuracy_score(y_pred , y_test) #'lilkel test tutarliligi'
```

```python
#Model Tunning
```

```python
mlp_params = {'alpha':[1,50.1,0.01,0.03],
              'hidden_layer_sizes':[(10,10),(100,100,100),(100,100),(3,5)],
             }
mlpc_model = MLPClassifier(solver = 'lbfgs')
```

```python
mlpc_cv = GridSearchCV(mlpc_model,mlp_params,cv = 10,n_jobs = -1,verbose = 2).fit(X_train,y_train)
```

```python
mlpc_cv.best_params_
```

```python
mlpc_tuned_model = mlpc_model = MLPClassifier(solver = 'lbfgs',activation = 'logistic',alpha = 1,hidden_layer_sizes=(100,100,100)).fit(X_train,y_train)
```

```python
y_pred = mlpc_tuned_model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```
