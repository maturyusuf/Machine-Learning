```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import model_selection
from matplotlib import pyplot as plt
from sklearn.linear_model import LassoCV
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv("/content/Hitters.csv")

# Drop the NaN values
df.dropna(inplace=True)
df.head()
```

```python
# ONE-HOT-ENCODING for non numerical data
dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])
dms.head()
```

```python
# Dropping the non numerical values and also ... ?
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis=1).astype("float64")
```

```python
X_.head()
```

```python
# Concatenating some* of  the one hot encoded features one hot encoded values with the dataframe.But why not ALL ?
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis=1)
```

```python
# Label
y = df["Salary"]
```

```python
# Holdout - Train and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#  First Model

model = Lasso().fit(X_train, y_train)
dir(model)
```

```python
coefs = []
alphas = np.random.randint(0,1000,10)
for i in alphas:
  model.set_params(alpha=i)
  model.fit(X_train, y_train)
  coefs.append(model.coef_)
```

```python
coefs # different weights for different lambda values
```

```python
# Prediction
```

```python
y_pred=model.predict(X_test)
pred_df=pd.DataFrame(y_pred, columns=["Predicted"], dtype="float64")
pred_df.reset_index(drop=True, inplace=True)

y_test_act=pd.DataFrame(y_test, columns=["Actual"],dtype="float64")
y_test_act.reset_index(drop=True, inplace=True)
new_df=pd.concat([y_test_act,pred_df ],axis=1)
new_df
```

```python
my_arr=np.array([y_test,y_pred]).T
my_arr.shape
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# Tunning Model

lasso_cv = LassoCV(cv=10, max_iter=100000).fit(X_train, y_train)
lasso_cv.alpha_
```

```python
lasso_tuned=Lasso().set_params(alpha = lasso_cv.alpha_).fit(X_train, y_train)
```

```python
np.sqrt(mean_squared_error(lasso_tuned.predict(X_test), y_test))
```

```python
alphas = 10**np.linspace(10,-2, 100)*0.5

lasso_cv = LassoCV(alphas=alphas,cv=10, max_iter=100000).fit(X_train, y_train)

lasso_tuned=Lasso().set_params(alpha = lasso_cv.alpha_).fit(X_train, y_train)


pd.Series(lasso_tuned.coef_, index = X_train.columns) # Check the 0 ones !!!
```

```python
np.sqrt(mean_squared_error(lasso_tuned.predict(X_test), y_test))
```
